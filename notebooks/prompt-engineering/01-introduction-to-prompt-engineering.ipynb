# Introduction to Prompt Engineering

Source slide: `slides/prompt-engineering/01_Introduction_to_Prompt_Engineering.md`

Prompt engineering becomes useful when an LLM has a broad goal but lacks the task definition, audience, constraints, or success criteria needed to respond reliably.


## Why this topic matters

### Generic goals create generic outputs

When a request only states the general intent, the model fills in missing assumptions on its own. That often produces bland responses, the wrong audience, or content that cannot be used directly.


```python
failure_example = {
    "prompt": 'Help me write a launch announcement for our new analytics product.',
    "likely_output": 'We are excited to announce our amazing new analytics product. It helps teams move faster with modern dashboards and insights.',
    "why_it_fails": 'The output is generic because the model was never told who the audience is, what differentiators matter, or what format success looks like.',
}

print('Failure example')
for key, value in failure_example.items():
    label = key.replace('_', ' ').title()
    print(f'\n{label}:\n{value}')
```

## Technique

### Technique: define the job before asking for the output

Prompt engineering starts by making the task concrete: describe the audience, goal, format, constraints, and any must-use context so the model can optimize toward a real target.


```python
improved_example = {
    "prompt": 'Write a 120-word product launch announcement for existing enterprise customers. Highlight SOC 2 readiness, row-level security, and faster dashboard performance. Use a confident but practical tone and end with a call to book a demo.',
    "likely_output": 'Introducing our new analytics platform for enterprise teams that need trusted speed at scale. It combines SOC 2-ready controls, row-level security, and faster dashboard performance so analysts can share insights without compromising governance. Teams can move from raw data to executive reporting with less manual work and stronger confidence in access controls. If you are planning your next analytics rollout, book a demo to see how the platform fits your environment.',
    "why_it_works": 'The improved prompt defines audience, differentiators, length, tone, and closing action, so the model has enough context to produce something usable on the first pass.',
}

print('Technique example')
for key, value in improved_example.items():
    label = key.replace('_', ' ').title()
    print(f'\n{label}:\n{value}')
```
