# Resources for Further Learning

Source slide: `slides/prompt-engineering/09_Resources_for_Further_Learning.md`

Further-learning resources are useful because prompt engineering skill degrades when teams rely only on intuition instead of studying working patterns, research, and hands-on examples.


## Why this topic matters

### Without references, teams repeat the same prompt mistakes

LLM behavior changes across tasks and models. If you never compare your prompts against proven patterns or published techniques, you can get stuck with brittle habits.


```python
failure_example = {
    "prompt": 'Summarize this research paper for executives.',
    "likely_output": 'This paper is about AI research and has several interesting findings relevant to business leaders.',
    "why_it_fails": 'The request is too underspecified, and without an external rubric or examples, the team has no benchmark for what a good executive summary prompt should include.',
}

print('Failure example')
for key, value in failure_example.items():
    label = key.replace('_', ' ').title()
    print(f'\n{label}:\n{value}')
```

## Technique

### Technique: use curated resources to upgrade your prompt patterns

Study prompt guides, notebooks, and research papers to collect reusable structures, evaluation rubrics, and examples. Then bring those learnings back into your prompts and experiments.


```python
improved_example = {
    "prompt": 'After reviewing a prompt guide that recommends audience, decision relevance, and evidence-backed bullets, rewrite the request as: “Summarize this research paper for executives in three bullets: business implication, supporting evidence, and decision relevance. Quote one finding directly from the paper.”',
    "likely_output": '- Business implication: The paper shows retrieval quality materially affects answer accuracy in production assistants.\n- Supporting evidence: “Answer correctness improved by 18% when citations were included.”\n- Decision relevance: Teams should invest in retrieval evaluation before scaling assistant usage.',
    "why_it_works": 'The improved prompt borrows a stronger pattern from external learning resources, turning vague intuition into a reusable prompt template.',
}

print('Technique example')
for key, value in improved_example.items():
    label = key.replace('_', ' ').title()
    print(f'\n{label}:\n{value}')
```

## Try it yourself

Adapt the improved prompt to one of your own tasks. Keep the same technique, then compare the first draft output against the improved version to see what changed.
